```python
from PIL import Image, ImageEnhance, ImageDraw
import numpy as np
from IPython.display import display, Image as IPImage

def apply_sunglasses_filter(image_path, output_path):
    # Open an image file
    with Image.open(image_path) as img:
        # Darken the image
        enhancer = ImageEnhance.Brightness(img)
        img = enhancer.enhance(0.5)  # Adjust brightness

        # Apply a gradient to simulate sunglasses
        gradient = Image.new('L', img.size, 0)
        draw = ImageDraw.Draw(gradient)
        draw.rectangle([0, 0, img.width, img.height // 2], fill=0)
        draw.rectangle([0, img.height // 2, img.width, img.height], fill=100)
        
        gradient = gradient.convert('RGB')
        img = Image.blend(img, gradient, alpha=0.4)  # Combine images

        # Save the filtered image
        img.save(output_path)

        # Display the filtered image
        display(IPImage(filename=output_path))

# Apply the filter to the uploaded image and display the result
apply_sunglasses_filter('IMG_20240519_101452.jpg', 'sunglasses_filter_image.jpg')
```
